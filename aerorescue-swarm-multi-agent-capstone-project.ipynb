### 🎯 Capstone Requirements Demonstrated
1. **Multi-agent systems:** 3 distinct agents (Recon, Medical, Commander) orchestrating logistics.
2. **MCP servers:** FastMCP mock server supplying real-time simulated disaster telemetry.
3. **Agent Skills:** Autonomous triage prioritization and route sorting.

### 📝 Project Rationale
In disaster zones, delayed logistics and poor triage data cost lives. Centralized, human-led dispatch is too slow for dynamic emergency environments. AeroRescue Swarm is a decentralized Python-based AI architecture that eliminates manual decision-making bottlenecks by processing environmental telemetry and health vitals to orchestrate drone-based supply drops to the highest-priority sectors.
